```python
fuel_cost = 450
CO2_cost = 103
reputation = 0.8


y_class = 0
j_class = 0
f_class = 61

y_price = 2790
j_price = 4830
f_price = 5220

AC_fuel_use = 9.4
AC_CO2 = 0.12

route_distance = 3106
```

```python
flight_revenue = (y_class*y_price)+(j_class*j_price)+(f_class*f_price)
fuel_consumption = AC_fuel_use*route_distance
fuel_cost_for_flight = (fuel_cost)*(fuel_consumption/1000)
quota_use = AC_CO2*((y_class*reputation)+((j_class*2)*reputation)+((f_class*3)*reputation))*route_distance
quota_cost_for_flight = (CO2_cost)*(quota_use/1000)
flight_profit = flight_revenue-(fuel_cost_for_flight+quota_cost_for_flight)

print("Flight Fuel Cost: " + "$" + str(int(fuel_cost_for_flight)))
print("Flight Co2 Quota Cost: " + "$" + str(int(quota_cost_for_flight)))
print("Flight Revenue: " + "$" + str(int(flight_revenue)))
print("Flight Profit: " + "$" + str(int(flight_profit)))
print("quotas:" + str(int(quota_use)))
```
